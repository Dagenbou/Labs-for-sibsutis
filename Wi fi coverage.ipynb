```python
import matplotlib.pyplot as plt 
figure, axes = plt.subplots() 
cc = plt.Circle(( 0 , 0 ), 44 , fill = False)
cc1 = plt.Circle(( 0 , 32 ), 32 , fill = False)
axes.set_aspect( 1 ) 
plt.xlim( -100 , 100 ) 
plt.ylim( -100 , 100 )
axes.add_artist( cc )
axes.add_artist( cc1 )
plt.title( 'Coverage' ) 
plt.show()
```
